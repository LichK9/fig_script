```python
import numpy as np    
import matplotlib.pyplot as plt    
import openpyxl

#bar patterns
patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*" ]

N=15
x=np.arange(N) 
xclicks=[
    '605.mcf_s','619.lbm_s','620.omnetpp_s','625.x264_s','638.imagick_s',
    'degreeCentr','graphColoring','kCore','pageRank','triangleCount',
    '433.milc','444.namd','462.libquantum','473.astar','avg'
]
xlabels=''
ylabels='Norm. Bandwidth(%)'
wb = openpyxl.load_workbook(filename="bandwidth.xlsx")
ws = wb['Sheet1']

b1=np.array([])

for col in ws.iter_cols(min_row=2,min_col=2,values_only=True):
    for value in col:
        # 获取单元格数据
        b1=np.append(b1,value * 100)     
        print(value)            
b1=b1.reshape(2, 15)
print(b1)


allfigsize=(10,3)     
width=0.4
plt.figure(figsize=allfigsize)

print(b1[0])
plt.bar(x - width*0.5, b1[0], width, label='CXL IDE & CTR', color='#082a54', edgecolor='black', linewidth=0.5, align='center')
plt.bar(x + width*0.5, b1[1], width, label='Adaptive Encryption', color='#f0c571',  hatch="//", edgecolor='black', linewidth=0.5, align='center')

plt.grid(axis='y',linestyle='--')
plt.xticks(x,xclicks, fontsize=15, rotation=30,ha='right', va='top') 
#plt.ylim(0,180000)
plt.yticks([50,100],fontsize=13)
#plt.xlabel(xlabels,fontsize=15)
plt.ylabel(ylabels,fontsize=15)    
plt.legend(loc='upper center', ncol=3, bbox_to_anchor=(0.50, 1.20), fontsize=13)    
plt.savefig('bandwidth.pdf',bbox_inches = 'tight')    
plt.show()
```
