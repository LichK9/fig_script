```python
import numpy as np    
import matplotlib.pyplot as plt    
import openpyxl

#bar patterns
patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*" ]

N=15
x=np.arange(N) 
xclicks=[
    '605.mcf_s','619.lbm_s','620.omnetpp_s','625.x264_s','638.imagick_s',
    'degreeCentr','graphColoring','kCore','pageRank','triangleCount',
    '433.milc','444.namd','473.astar','462.libquantum','avg'
]
xlabels=''
ylabels='Norm. Exec. Time'
wb = openpyxl.load_workbook(filename="introduction.xlsx")
ws = wb['Sheet1']

b1=np.array([])

for col in ws.iter_cols(min_row=2,min_col=2,values_only=True):
    for value in col:
        # 
        b1=np.append(b1,value)     
        print(value)            
b1=b1.reshape(4, 15)
print(b1)


allfigsize=(10,2.5)     
width=0.15
plt.figure(figsize=allfigsize)
print(b1[0])
plt.bar(x - width*1.5, b1[0], width, label='Local Insecure', color='#cecece', hatch=".", edgecolor='black', linewidth=0.25) 
plt.bar(x - width*0.5, b1[2], width, label='Local XTS', color='#a559aa', hatch="-", edgecolor='black', linewidth=0.25)
plt.bar(x + width*0.5, b1[1], width, label='CXL Insecure', color='#59a89c', hatch="x", edgecolor='black', linewidth=0.25)
plt.bar(x + width*1.5, b1[3], width, label='CXL IDE & XTS', color='#e02b35', hatch="o", edgecolor='black', linewidth=0.25)

plt.grid(axis='y',linestyle='--')
plt.xticks(x,xclicks, fontsize=15, rotation=30,ha='right', va='center_baseline') 
#plt.ylim(0,180000)
plt.yticks([1.0,1.5,2.0],fontsize=13)
#plt.xlabel(xlabels,fontsize=15)
plt.ylabel(ylabels,fontsize=15)    
plt.legend(loc='upper center', ncol=4, bbox_to_anchor=(0.50, 1.23), fontsize=13)    
plt.savefig('introduction.pdf',bbox_inches = 'tight')    
plt.show()
```
